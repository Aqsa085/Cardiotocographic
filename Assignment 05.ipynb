## Detect and Treat outliers

```python
# Plot boxplot before treating outliers
plt.figure(figsize=(10,8))
df.boxplot()
plt.title('Boxplot Before Handling Outliers')
plt.show()





```

```python
def treat_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    IQR = q3 - q1

    # Calculate the upper and lower limits for outliers
    upper_limit = q3 + 1.5 * IQR
    lower_limit = q1 - 1.5 * IQR

    # Clip values to the limits
    df[column] = df[column].clip(lower=lower_limit, upper=upper_limit)

    return df
```

```python
# Detect and treat outliers for each column
for column in df.columns:
    df = treat_outliers(df, column)
```

```python
# Plot boxplot after treating outliers
plt.figure(figsize=(10,8))
df.boxplot()
plt.title('Boxplot After Treating Outliers')
plt.show()
```

## Interesting Findings:
Median vs. Mean:
If the mean and median differ significantly in any column, it suggests the data might be skewed.

IQR (Interquartile Range):
A large IQR indicates wide data spread. Columns with high IQR values may have more variability.
A small IQR suggests tightly packed data with little variation.

Outliers:
Columns where the mean is far from the 25th and 75th percentiles could indicate outliers pulling the mean.

The dataset contains 2,103 rows of fetal heart health data.
The central tendency indicates that the data falls within a very narrow range.
Out of 14 columns, 10 have a mean value greater than 10.
Most columns show a narrow interquartile range (IQR).
9 columns have an IQR smaller than 1.
4 columns have an IQR smaller than 30.
The baseline fetal heart rate (LB) is typically around 133 beats per minute.
The medians of the columns are generally close to zero.

```python
# Create pair plots
sns.pairplot(df)
plt.suptitle('Pair Plot of All Variables', y=1.04)
plt.show()
```

```python
#Create Violin Plots
# Get the total number of columns (variables) in the DataFrame
num_vars = len(df.columns)

# Set the number of columns for the plot grid
num_cols = 4

# Calculate how many rows we need based on the number of variables
num_rows = (num_vars + num_cols - 1) // num_cols

# Set the size of the overall figure (adjusting height based on rows)
plt.figure(figsize=(12, 6* num_rows))

# Loop through each column and create a violin plot for it
for i, column in enumerate(df.columns):
    plt.subplot(num_rows, num_cols, i + 1)  # Create subplot for each variable
    sns.violinplot(y=df[column])  # Create a violin plot for the current variable
    plt.title(column)  # Set the title as the column name

plt.tight_layout()  # Adjust layout to avoid overlapping
plt.show()  # Display the plots
```
